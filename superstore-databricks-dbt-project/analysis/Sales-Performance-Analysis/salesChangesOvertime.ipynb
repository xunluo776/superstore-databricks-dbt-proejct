```sql
SELECT
    Row_ID,
    Product_Name,
    Sales,
    Profit,
    Discount,
    Customer_Name,
    Order_ID
FROM superstore_databricks_dbt.bronze.superstore

ORDER BY Row_ID;
```

```sql
SELECT
  date_format(
    CAST(order_date AS DATE),
    'yyyy-MM'
  ) AS year_month,
  SUM(
    try_cast(sales AS DOUBLE)
  ) AS monthly_sales
FROM superstore_databricks_dbt.dbt_luo_gold.superstore_fact
GROUP BY year_month
```

```sql
%python
from pyspark.sql import SparkSession
```

```sql
%python


spark = SparkSession.builder.appName("SuperstoreFix").getOrCreate()

df = (
    spark.read
        .format("csv")
        .option("header", "true")
        .option("multiLine", "true")        # handle multi-line text
        .option("quote", '"')              # treat double quotes as text wrapper
        .option("escape", '"')             # handle embedded quotes like 9 1/2"" x 11""
        .option("inferSchema", "true")
        .option("ignoreLeadingWhiteSpace", "true")
        .option("ignoreTrailingWhiteSpace", "true")
        .load("/Volumes/superstore_databricks_dbt/source/raw/superstore/")
)

df.show(5, False)
```
